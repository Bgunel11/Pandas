```python
# Əgər ödənən hesab 10 -dan azdırsa $ simvolu, 10 və 30 arasındadırsa $$, 30 dan çoxdursa $$$ simvolu yeni sütunda uyğun olaraq yazılsın
```

# I yol

```python
df.head()
```

```python
def fiyat_kateqori(x):
    if x < 10 :
        return '$'
    elif 10 <= x < 30 :
        return '$$'
    else :
        return '$$$'



```

```python
df['total_bill'].apply(fiyat_kateqori)
```

```python
df.head()
```

```python
df['price_category'] = df['total_bill'].apply(fiyat_kateqori)
```

```python
df.head()
```

# II yol lambda expression

```python
df.head()
```

```python
df['price_category'] = df['total_bill'].apply(lambda x : '$' if x < 10 else '$$' if 10 <= x < 30 else '$$$')
```

```python
df.sample(10)
```

# III yol vectorized (ən sürətlisi)

```python
np.full((2,3), 4)
```

```python
def fiyat_kateqori_vectorized(bill):
    categories = np.full_like(bill, '$', dtype= 'U3')
    categories[(bill >= 10) & (bill < 30)] = '$$'
    categories[bill >= 30] = '$$$'
    return categories
```

```python
df['total_bill'].keys()
```

```python
df['total_bill'].values
```

```python
df['price_category'] = fiyat_kateqori_vectorized(df['total_bill'].values)
```

```python
df.sample(7)
```

# Between() methodu

```python
df[(df['total_bill'] > 10) & (df['total_bill'] < 20)]
```

```python
df[df['total_bill'].between(10, 20)] # default olaraq ikisi de daxildir.
```

```python
df[df['total_bill'].between(10, 20 , inclusive= 'both')] # hər ikisi daxil olmaqla
```

```python
df[df['total_bill'].between(10, 20, inclusive='neither')] # heç biri daxil deyil
```

```python
df[df['total_bill'].between(10, 20, inclusive='left')] #yalnız sol sərhəd daxil edilir
```

```python
df[df['total_bill'].between(10, 20, inclusive='right')] #yalnız sağ sərhəd daxildir.
```

```python
df[~df['total_bill'].between(10, 20)]
```
