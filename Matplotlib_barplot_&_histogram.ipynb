```python
data = {
    "Year": np.arange(2010, 2021),
    "Sales": np.random.randint(100, 500, 11),
    "Payment": np.random.randint(50, 400, 11)
}

df = pd.DataFrame(data)
df["Region"] = np.random.choice(["south", "north","east","west"], df.shape[0])
df
```

```python
plt.figure(figsize= (8, 7))
plt.plot(df["Year"], df["Sales"], label = "Sales for Years",
         marker = "o", markersize = 12, markerfacecolor = "black",
         linestyle= "--", linewidth = 3, color = "blue")

plt.plot(df["Year"], df['Payment'], label = "Payment for Years",
         marker= "s", markersize= 12, markerfacecolor="green",
         linestyle="-", linewidth = 3, color ="red")

plt.title("illik satış və xərclər", fontsize= 16)
plt.xlabel("years", fontsize= 14)
plt.ylabel("AZN", fontsize= 14)


plt.legend(loc= "best")
plt.show()
```

```python
sales_by_region = df.groupby("Region")["Sales"].sum()
sales_by_region
```

```python
sales_by_region.index
```

```python
sales_by_region.values
```

```python
plt.figure(figsize=(8, 6))
plt.bar(sales_by_region.index, sales_by_region.values, color= "red", width= 0.5,
        edgecolor= "green", linewidth = 4, label="ümumi satış")

plt.legend(loc="best")

plt.title("Hər region üzrə ümumi satış", fontsize=20)
plt.xlabel("Region", fontsize=15)
plt.ylabel("AZN", fontsize= 15)

plt.show()

```
