# Giriş

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize= (12, 8))
plt.show()
```

# plt.plot() funksiyası



```python
# f(x) = 2x

x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

plt.figure(figsize= (6,4))
plt.plot(x, y);
plt.show()
```

```python
# f(x) = -3x

x = [0, 1, 2, 3, 4, 5]
y = [0, -3, -6, -9, - 12, - 15]

plt.figure(figsize=(6,4))
plt.plot(x, y, "--", color= "black")
plt.show()
```

```python
# 2 * np.arange(6) + 1
```

```python
x = np.arange(6)
y = 2*x + 1
# f(x) = y = 2*x + 1

plt.figure(figsize= (6, 5))
plt.plot(x, y, "r--")
plt.show()
```

```python
plt.figure(figsize= (6, 5))
plt.plot(x, y, "--", color= "green")
plt.show()
```

## Sual: `2x + 3` funksiyasını numpy istifadə edin və plt.plot ilə qrafikini vizuallaşdırın.

```python
x = np.arange(4)
y = 2 * x + 3

plt.figure(figsize= (6, 4))
plt.plot(x, y, "--", color= "black")
plt.show()
```

# Linear olmayan funksiyalar

```python
x = np.arange(100) / 10
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(10)
y1 = x*2
y2 = x**2

plt.figure(figsize=(7, 6))

plt.plot(x, y1,"--", label= "2*x", color="red")
plt.plot(x, y2,"--", label= "x**2", color = "green")
plt.legend(loc="best")
plt.show()


```

# plt.scatter - səpələnmə funksiyası yaratmaq üçün

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.figure(figsize= (6, 5))

plt.plot(x, y, color="blue")
plt.scatter(x, y, color= "red")

plt.show()
```

```python
x = np.arange(100) / 10
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(7, 6))

plt.plot(x, y1, color="red", label= "sinx")
plt.scatter(x, y2, color="blue", label="cosx")

plt.legend(loc= "upper right")

plt.show()
```

# plt.plot parametrləri

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

plt.figure(figsize= (7, 5))

plt.plot(x, y, marker ="s",  linestyle = "-.", linewidth = 4, markersize = 15, color= "blue", alpha = 0.3, label="2x")

plt.legend(loc="best")

plt.show()
```

# plt.scatter parametrləri


```python
x = np.arange(100) / 10

plt.figure(figsize=(7, 6))

plt.scatter(x, np.cos(x), color="red", s=70, edgecolors ="blue", linewidths=3, alpha=0.4, label= "cosx")

plt.legend(loc="upper right")

plt.show()
```
